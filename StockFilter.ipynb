```python
import os,sqlite3,csv
import datetime,time
from datetime import  timedelta
end=datetime.date.today()
start=end- timedelta(days=30)
print(end,start)
conn = sqlite3.connect('stock.db')
c = conn.cursor()
# 選出所有股票
stocklist=[]
for stockid in c.execute("select distinct ID from StockDaily"):
    stocklist.append(stockid[0])
count=0
for stockid in stocklist:
    c.execute("select TradeDate,ID,Amount,Close from StockDaily where TradeDate between " + start.strftime("%Y%m%d")+" and " + end.strftime("%Y%m%d") +" and ID='"+stockid+"' order by TradeDate")
    CloseList=c.fetchall()
    close=[]
    if(len(CloseList)>0):
        for price in CloseList:
            #print(price)
            close.append(price[3])
        if(min(close)==0):
            continue
        maxClose=max(close)
        minClose=min(close)
        maxCloseIndex=close.index(max(close))
        minCloseIndex=close.index(min(close))
        currentClose=close[len(close)-1]
        #print(maxClose,minClose)
        maxDate=CloseList[maxCloseIndex][0]
        minDate=CloseList[minCloseIndex][0]
        if(maxDate>minDate):
            firstM=(maxClose-minClose)/(maxCloseIndex-minCloseIndex+1)
            
            print(stockid,maxDate,maxClose,minDate,minClose,currentClose,firstM)
            count+=1
print(count)
    
```
